```python
# Dependencies
import requests
import json
import pandas as pd
from pandas import DataFrame
from pandas.io.json import json_normalize
```

```python
# attempt working with file names
for p in range (50):
    url = "https://www.rijksmuseum.nl/api/nl/collection?key=SlCXqeYu&format=json"
    page = f"&p={p}"
    page_number = f"page_{p}"
    print(page_number)
    page_query = url + page
    print(page_query)
    
    response = json.dumps(requests.get(page_query).json(), indent=4, sort_keys=True)

    
    writeFile = open(f"page_{p}.json", 'w')
    writeFile.write(response)
    writeFile.close()
    print("API call "f"{p}"" completed")
    print("===================================")
    page = p+1
```
